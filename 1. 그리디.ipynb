## 1. 그리디
- 그리디 알고리즘(탐욕법)은 현재 상황에서 지금 당장 좋은 것만 고르는 방법
- 일반적인 그리디 알고리즘은 문제를 풀기 위한 최소한의 아이디어를 떠올릴 수 있는 능력을 요구함.
- 그리디 해법은 그 정당성 분석이 중요하다.
    * 단순히 가장 좋아보이는 것을 반복적으로 선택해도 최적의 해를 구할 수 있는지 검토해야함.
- 일반적인 상황에서는 최적의 해를 보장할 수 없을 때가 많다.
- 코딩테스트에서 그리디 문제는 최적의 해를 보장함을 추론할 수 있도록 출제됨.

[거스름돈]

```python
n=int(input())
array=[500, 100, 50,10]
count=0

for coin in array:
    count+=n//coin
    n%=coin
print(count)
```

[1이 될 때까지]


```python
n,k=map(int,input().split())
result=0
while True:
    target = (n//k)*k
    result +=(n-target)#1을 빼는 과정 수
    n=target
    if n<k:
        break
    result += 1#k로 나누는 과정 수
    n//=k
result +=(n-1)
print(result)

```

[곱하기 혹은 더하기]


```python
s=input()
result =int(s[0])

for i in range(len(s)-1):
    if int(s[i])<=1 or int(s[i+1])<=1:
        result += int(s[i+1])
    else:
        result *=int(s[i+1])
        
print(result)
        
```

[모험가길드]


```python
n=int(input())
data = list(map(int,input().split()))
data.sort()

result=0
count=0

for i in data:
    count+=1
    if count >=i:
        result+=1
        count=0
print(result)
```
