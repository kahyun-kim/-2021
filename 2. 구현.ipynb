### 알고리즘에서의 2차원 공간


[상하좌우]


```python
#시작 = 젤왼위(1,1), 젤오아래(N,N)
#L R U D
n=int(input())
direction=input().split()
x,y=1,1

dx=[-1,1,0,0]
dy=[0,0,-1,1]
command=['L','R','U','D']

for d in direction:
    for i in range(len(command)):
        if d==command[i]:
            nx=x+dx[i]
            ny=y+dy[i]
    if nx<1 or ny<1 or nx>n or ny>n:
        continue
    x,y=nx,ny
    
print(y,x)
    
```

[왕실의 나이트]
- c++의 경우 dx,dy 나누는게 더 편함
- 8가지 방향을 모두 dx, dy로 표현 가능

***나이트가 움직일 수 있는 방향***  
*1. 수평으로 두 칸 이동한 뒤에 수직으로 한 칸 이동하기*  
*2. 수직으로 두 칸 이동한 뒤에 수평으로 한 칸 이동하기*




```python
position=input()#a1 의 문자열, a~h, 1~8

temp=['a','b','c','d','e','f','g','h']
x=int(temp.index(position[0]))+1
y=int(position[1])

dx=[-2,2,0,0]
dy=[0,0,-2,2]
direction=['L','R','U','D']
count=8

for i in direction:
    nx1=x+dx[direction.index(i)]
    ny1=y+dy[direction.index(i)]
    nx2=x+dx[direction.index(i)]
    ny2=y+dy[direction.index(i)]
    if i=='L' or i=='R':
        ny1+=1
        ny2-=1
    elif i=='U' or i=='D':
        nx1+=1
        nx2-=1
    if nx1<1 or nx1>8 or ny1<1 or ny1>8:
        count-=1
    if nx2<1 or nx2>8 or ny2<1 or ny2>8: 
        count-=1
        
print(count)
```

- 해답

```python
input_data=input()
row = int(input_data[1])
column=int(ord(input_data[0]))-int(ord('a'))+1 #아스키코드 값 차이로 가져옴

#8가지 경로에 대해서 dx,dy 두개 말고 하나의 리스트로 저장
steps=[(-2,-1),(-1,-2),(1,-2),(2,-1),(2,1),(1,2),(-1,2),(-2,1)]

result=0
for step in steps:
    next_row = row + step[0]
    next_column = column + step[1]
    if next_row>=1 and next_row <=8 and next_column>=1 and next_column<=8:
        result+=1
        
print(result)
```

[문자열 재정렬]
- 수 = 정수
- 숫자 = 글자
- string.isalpha() = 알파벳이면 true 출력



```python
s=input()
temp_sum=0
temp_text=list()

for i in range(len(s)):
    if int(s[i]) >= 1 and int(s[i])<= 9 :
        temp_sum+=int(s[i])
    elif ord(s[i]) >=ord('A') and ord(s[i])<=ord('Z'):
        temp_text.append(s[i])
temp_text.sort()
temp_text.append(temp_sum)
print(''.join(temp_text))
```

- 해답

```python
s=input()
temp_sum=0
temp_text=[]

for i in s:
    if i.isalpha():
        temp_text.append(i)
    else:
        temp_sum+=int(i)
temp_text.sort()
if temp_sum !=0:
    temp_text.append(str(temp_sum))
print(''.join(temp_text))
```
